```python
def reverse_complement(seq: str) -> str:
    complement = {
        "A": "T",
        "T": "A",
        "C": "G",
        "G": "C",
        "N": "N"
    }
    
    seq = seq.upper()
    rev = seq[::-1]
    
    result = []
    for base in rev:
        if base not in complement:
            raise ValueError(f"Invalid nucleotide: {base}")
        result.append(complement[base])
    
    return "".join(result)
```

```python
reverse_complement("ATGC")
```

```python
def test_reverse_complement():
    assert reverse_complement("ATGC") == "GCAT"
    assert reverse_complement("atgc") == "GCAT"   # case insensitivity
    assert reverse_complement("AANNTT") == "AANNTT"  # ambiguity code N

    print("All tests passed!")
```

```python
test_reverse_complement()
```

```python
def read_fasta(filepath: str):
    header = None
    sequence_parts = []
    
    with open(filepath, "r") as f:
        for line in f:
            line = line.strip()
            if not line:
                continue
            if line.startswith(">"):
                header = line
            else:
                sequence_parts.append(line)
    
    sequence = "".join(sequence_parts)
    return header, sequence
```

```python
header, seq = read_fasta("lambda_phage.fasta")
print(header)
print(len(seq))
```
